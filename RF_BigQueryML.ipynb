```python
import pandas_gbq
import pandas as pd
from sklearn.model_selection import train_test_split
project_id = 'datamining-415802'
```

```python
df = pd.read_csv('./diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
df
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)
```

```python
# Merge back into one dataframe for EDA
diabetes_train = pd.DataFrame(x_train)
diabetes_train.insert(0, "Diabetes_binary", y_train)
diabetes_train.head()
```

```python
# Merge back into one dataframe for EDA
diabetes_test = pd.DataFrame(x_test)
diabetes_test.insert(0, "Diabetes_binary", y_test)
diabetes_test.head()
```

```python
pandas_gbq.to_gbq(diabetes_train, 'my_dataset.diabetes_train', project_id=project_id)
```

```python
pandas_gbq.to_gbq(diabetes_test, 'my_dataset.diabetes_test', project_id=project_id)
```
